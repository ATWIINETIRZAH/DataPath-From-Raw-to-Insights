```python
#import relevant libraries

#for data manipulation
import pandas as pd
#for numerical operations
import numpy as np
#for statistical visualization
import seaborn as sns
#library used for data plotting
import matplotlib.pyplot as plt
```

```python
#create a dataframe named grades and use it to read the dataset(excel file-student performance)
grades= pd.read_excel('student_performance.xlsx')

grades.head(4)
```

```python

#Plotting a pie chart for a 'Category' column against its 'Values'
grades_grouped = grades.groupby('Age')['DSC3108'].sum()

# Create the pie chart
plt.pie(grades_grouped, labels=grades_grouped.index, autopct='%1.2f%%', startangle=90)
plt.title('Pie Chart of age and performance in DSC3108 ')
plt.show()
```

```python

#Determine the size of the line graph
plt.figure(figsize=(10, 7))

#loops through the rows to get all students marks in the courses
for index, row in grades.iterrows():
    plt.plot(
        ['DSC3108', 'DSC3114', 'SYE3104', 'MTH3108'], 
        [row['DSC3108'], row['DSC3114'], row['SYE3104'], row['MTH3108']], 
        label=f'Student {row["Access No"]}'
    )

#label fo x axis
plt.xlabel('Subjects')
# label for y axis
plt.ylabel('Scores')
#label fo rgraph title
plt.title('Line Graph: Students\' Performance Across Subjects')
plt.legend(title='AccessNo')
plt.show()
```

```python

#calculates the average of the grades per course
average_scores = grades[['DSC3108', 'DSC3114', 'SYE3104', 'MTH3108']].mean()

#determines the size
plt.figure(figsize=(8, 6))
#plots the average scores with different column colours
average_scores.plot(kind='bar', color=['blue', 'green', 'orange', 'red'])
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.title('Bar Graph: Average Scores per Subject')
plt.xticks(rotation=0)
plt.show()
```
