```python
#import relevant libraries

#for data manipulation
import pandas as pd
#for numerical operations
import numpy as np
#for statistical visualization
import seaborn as sns
#library used for data plotting
import matplotlib.pyplot as plt
```

```python
#create a dataframe named grades and use it to read the dataset(excel file-student performance)
grades= pd.read_excel('student_performance.xlsx')

grades.head(4)
```

```python

#Plotting a pie chart for a 'Category' column against its 'Values'
grades_grouped = grades.groupby('Age')['DSC3108'].sum()

# Create the pie chart
plt.pie(grades_grouped, labels=grades_grouped.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of age and performance in DSC3108 ')
plt.show()
```
